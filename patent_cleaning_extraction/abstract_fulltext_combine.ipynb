```python
for folder in os.listdir('fulltextrawdata2023'):
    print(f"doing {folder}")
    if folder == '.DS_Store':
        continue
    temp_df = pd.read_csv(f'fulltextrawdata2023/{folder}/individuals_consolidated.csv')
    print(temp_df.shape)
    # find all unique app_ref_doc_number
    unique_app_ref_doc_number = temp_df['app_ref_doc_number'].unique()
    # filter the abstracts dataframe
    small_abstracts = abstracts[abstracts['app_ref_doc_number'].isin(unique_app_ref_doc_number)]
    # merge only the description_text to the abstracts dataframe
    small_abstracts = small_abstracts.merge(temp_df[['app_ref_doc_number', 'description_text']], on='app_ref_doc_number')
    print(small_abstracts.shape)
    small_abstracts.to_csv(f'weekly_combined/{folder}.csv', index=False)
    
```

```python
# Combine all the weekly_combined files into a single one
import os
import pandas as pd
all_files = []
for file in os.listdir('weekly_combined'):
    print(f"doing {file}")
    if file == '.DS_Store':
        continue
    all_files.append(pd.read_csv(f'weekly_combined/{file}'))
#concat and remove rows that are duplicate app_ref_doc_number
pd.concat(all_files).drop_duplicates('app_ref_doc_number').to_csv('allpatents_with_abstracts_and_full_text.csv', index=False)
```
